```python
import matplotlib.pyplot as plt

# Data
models = [
    "Neive Bias", "Logistic Regression", "AdaBoost",
    "Gradient Boosting", "Voting Regression",
    "Random Forest", "Decision Tree", "Ensemble Method"
]
accuracy_with_feature_fusion = [100] * len(models)
accuracy_without_feature_fusion = [89.67, 97.84, 99.13, 99.24, 99.42, 99.62, 99.64, 99.73]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(models, accuracy_without_feature_fusion, label="Accuracy (Without Feature Fusion)", marker='o')
plt.plot(models, accuracy_with_feature_fusion, label="Accuracy (With Feature Fusion)", marker='o')

# Labels and Title
plt.title("Model Accuracy Comparison", fontsize=16)
plt.xlabel("Models", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```
