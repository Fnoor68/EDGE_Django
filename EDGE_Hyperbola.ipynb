```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['lines.color'] = 'k'
mpl.rcParams['axes.prop_cycle'] = mpl.cycler('color', ['k'])
```

```python
x = np.linspace(-9, 9, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
```

```python
def axes():
    plt.axhline(0, alpha=.1)
    plt.axvline(0, alpha=.1)
```

```python
a = .3
axes()
plt.contour(x, y, (y**2 - 4*a*x), [0], colors='k')
plt.show()
```

```python
axes()
plt.contour(x, y, (y**2 - 4*a*x), [0], colors='k', alpha=.1)
# Focus.
plt.plot(a, 0, '.')
# Directrix.
plt.axvline(-a)
plt.show()
```

```python
a, b, c, d, e, f = 1, -2, 1, 2, 2, -10
assert b**2 - 4*a*c == 0
axes()
plt.contour(x, y,(a*x**2 + b*x*y + c*y**2 + d*x + e*y + f), [0], colors='k')
plt.show()
```

```python
a = 2.
b = 2.
axes()
plt.contour(x, y,(x**2/a**2 + y**2/b**2), [1], colors='k')
plt.show()
```

```python
a = 4.
b = 2.
axes()
plt.contour(x, y,(x**2/a**2 + y**2/b**2), [1], colors='k')
plt.show()
```

```python
axes()
plt.contour(x, y,(x**2/a**2 + y**2/b**2), [1], colors='k', alpha=.1)
# Eccentricity.
e = np.sqrt(1 - b**2/a**2)
# Foci.
plt.plot(a*e, 0, '.', -a*e, 0, '.')
# Directrices.
plt.axvline(a/e)
plt.axvline(-a/e)
plt.show()
```

```python
a, b, c, d, e, f = 4, -5, 2, 4, -3, -3
assert b**2 - 4*a*c < 0
axes()
plt.contour(x, y,(a*x**2 + b*x*y + c*y**2 + d*x + e*y + f), [0], colors='k')
plt.show()
```

```python
a = 2.
b = 1.
axes()
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='k')
plt.show()
```

```python
axes()
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='k', alpha=.1)
# Eccentricity.
e = np.sqrt(1 + b**2/a**2)
# Foci.
plt.plot(a*e, 0, '.', -a*e, 0, '.')
# Directrices.
plt.axvline(a/e)
plt.axvline(-a/e)
# Asymptotes.
plt.plot(x[0,:], b/a*x[0,:], '--')
plt.plot(x[0,:], -b/a*x[0,:], '--')
plt.show()
```

```python
a, b, c, d, e, f = 1, -3, 1, 1, 1, 1
assert b**2 - 4*a*c > 0
axes()
plt.contour(x, y,(a*x**2 + b*x*y + c*y**2 + d*x + e*y + f), [0], colors='k')
plt.show()
```
